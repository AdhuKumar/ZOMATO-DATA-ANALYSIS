```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("/Users/adhukumara/Desktop/ZOMATO DATA ANALYSIS/Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# convert the data type of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe
```

```python
dataframe.info()
```

# Type of Restaurant 

```python
dataframe.head()
```

```python
sns.countplot(x = dataframe['listed_in(type)'])
plt.xlabel('type of restaurant')
        
```

# conclusion - majority of the restaurant falls in dinning categories

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green",marker = "o")
plt.xlabel("Type of restaurant", c ="red" ,size=20)
plt.ylabel("votes", c ="red" ,size=20)
```

# conclusion - dinning restaurants has recieved maximum votes

# The Majority of restaurant received ratings

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins = 5)
plt.title("rating distribution")
plt.show()
```

# conclusion - the majority restaurant received rating from 3.5 to 4

# Average order spending by couples

```python
dataframe.head()
```

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

# conclusion - the majority of couples prefers restaurants with an approximate cost of 300 rupees

# which mode receives maximum rating

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,8))
sns.boxplot(x = 'online_order' , y = 'rate' ,data = dataframe)
```

# Conclusion - offline order received lower rating as compared to online orders

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)

# Create the heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')

# Add labels and title
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

# Conclusion - Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurants, but prefer online ordering at cafes.
